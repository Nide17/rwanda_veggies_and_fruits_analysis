```python
# IMPORTING LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# BRING IN THE .xlsx FILE FROM THE NATIONAL INSTITUTE OF STATISTICS OF RWANDA
df = pd.read_excel("2023 Season _Final Tables.xlsx", sheet_name="Table 5")
df.tail()
```

```python
# DROP THE FIRST COLUMN
df = df.drop(columns=["Unnamed: 0"])
df.tail()
```

```python
# USE ROW 27 AND 28 AS A NEW DATAFRAME
veggies_and_fruits = df.iloc[27:29]
veggies_and_fruits
```

```python
# 4 VARIABLES: Cultivated Crop area (Ha), Harvested Crop area (Ha), Production (MT), Yield (MT/ha): EACH HAS THESE 3 SEASONS OF DATA: 2023 B, 2022 B, 2021 B TO MAKE TITLE AS  variable season
vars = [
    "Cultivated Crop area (Ha)",
    "Harvested Crop area (Ha)",
    "Production (MT)",
    "Yield - Prod/Harv. area (MT/ha)",
]

seasons = ["2023 B", "2022 B", "2021 B"]
seasons_ordered = ["2021 B", "2022 B", "2023 B"]
seasons_years = ["2021", "2022", "2023"]

# CREATE A LIST OF TITLE NAMES
titles = []
for var in vars:
    for season in seasons:
        title = var + " - " + season
        titles.append(title)

# ADD Crop Name TITLE IN FRONT OF THE LIST OF TITLES
titles = ["Crop Name"] + titles
```

```python
# ADD THE TITLES TO THE veggies_and_fruits DATAFRAME
veggies_and_fruits.columns = titles

# MAKE THE CROP NAME THE INDEX
veggies_and_fruits = veggies_and_fruits.set_index("Crop Name")

# CHRONOLOGICAL ORDER
veggies_and_fruits = veggies_and_fruits[[titles[3], titles[2], titles[1], titles[6], titles[5], titles[4], titles[9], titles[8], titles[7], titles[12], titles[11], titles[10]]]
veggies_and_fruits
```

```python
# EXPLORATORY DATA ANALYSIS
df.info()
```

```python
# DESCRIBE THE PRODUCTION (MT) COLUMN OF THE veggies_and_fruits DATAFRAME FOR THE 2023 B SEASON 
veggies_and_fruits["Yield - Prod/Harv. area (MT/ha) - 2023 B"].describe()
```

```python
# PRODUCTION TREND ANALYSIS -  PLOTTING THE PRODUCTION OVER THE 3 SEASONS
veggies_and_fruits.iloc[:, 6:9].plot(
    kind="bar",
    figsize=(18, 7),
    width=0.8,
    linewidth=1,
    edgecolor="xkcd:dark blue",
    color=["xkcd:light blue", "xkcd:light green", "xkcd:light yellow"],
    fontsize=12,
    rot=0,
)

# background color
ax = plt.gca()
ax.set_facecolor("xkcd:light grey")
ax.set_alpha(0.2)

# title and labels
plt.title(
    "TREND ANALYSIS FOR PRODUCTION OVER 3 RECENT YEARS",
    size=20,
    pad=20,
    fontweight="bold",
    color="xkcd:dark blue",
)
plt.xlabel("Crop Name", size=15, labelpad=8, fontweight="bold", color="xkcd:dark blue")
plt.ylabel("Area (Ha)", size=15, labelpad=8, fontweight="bold", color="xkcd:dark blue")

# label individual bars with quantity
for i in range(0, 3):
    ax.bar_label(ax.containers[i], label_type="center", fontsize=12, color="xkcd:dark blue")

# legend
plt.legend(loc="upper right", fontsize=12)
plt.show()
```

```python
# ANALYSIS OF THE PRODUCTION (Production (MT)) FOR EACH SEASON - Vegetables and Fruits
fig, axes = plt.subplots(
    figsize=(20, 7),
    sharey=True,
    dpi=100,
    facecolor="xkcd:light grey",
    edgecolor="xkcd:dark blue",
)

x = np.arange(3)
width = 0.35

axes.bar(
    x,
    veggies_and_fruits.iloc[0, 6:9],
    width,
    label="VEGETABLES PRODUCTION (MT)",
    color="xkcd:light blue",
    linewidth=1,
    edgecolor="xkcd:dark blue",
)
axes.bar_label(axes.containers[0], label_type="center", color="xkcd:dark blue", fontsize=15)
axes.bar(
    x + width,
    veggies_and_fruits.iloc[1, 6:9],
    width,
    label="FRUITS PRODUCTION (MT)",
    color="xkcd:light pink",
    linewidth=1,
    edgecolor="xkcd:dark blue",
    animated=True,
    mouseover=True,
)
axes.bar_label(axes.containers[1], label_type="center", color="xkcd:dark blue", fontsize=15)

# title and labels
axes.set_xlabel(
    "Season", size=15, labelpad=8, fontweight="bold", color="xkcd:dark blue"
)
axes.set_ylabel(
    "Production (MT)", size=15, labelpad=8, fontweight="bold", color="xkcd:dark blue"
)
axes.set_title(
    "VEGETABLES AND FRUITS PRODUCTION COMPARISON PER YEAR (2021 - 2023)",
    size=13,
    pad=20,
    fontweight="bold",
    color="xkcd:dark blue",
)

axes.set_xticks(x)
axes.set_xticklabels(seasons_ordered)
axes.legend()

plt.show()
```

```python
# ANALYSIS OF THE AREAS (Cultivated Crop area (Ha), Harvested Crop area (Ha)) FOR EACH SEASON - Vegetables
fig, axes = plt.subplots(
    1,
    2,
    figsize=(20, 5),
    sharey=True,
    dpi=100,
    facecolor="xkcd:light grey",
    edgecolor="xkcd:dark blue",
)

x = np.arange(3)
width = 0.4

# CULTIVATED AREA (Ha)
axes[0].bar(
    x - width / 2,
    veggies_and_fruits.iloc[0, 0:3],
    width,
    label="CULTIVATED AREA (Ha)",
    color="xkcd:light blue",
    linewidth=1,
    edgecolor="xkcd:dark blue",
)
axes[0].bar_label(
    axes[0].containers[0], label_type="center", color="xkcd:dark blue", fontsize=12
)

# HARVESTED AREA (Ha)
axes[0].bar(
    x + width / 2,
    veggies_and_fruits.iloc[0, 3:6],
    width,
    label="HARVESTED AREA (Ha)",
    color="xkcd:light green",
    linewidth=1,
    edgecolor="xkcd:dark blue",
)
axes[0].bar_label(
    axes[0].containers[1], label_type="center", color="xkcd:dark blue", fontsize=12
)

axes[0].set_ylabel(
    "Area (Ha)", size=12, labelpad=8, fontweight="bold", color="xkcd:dark blue"
)
axes[0].set_title(
    "CULTIVATED AREA VS HARVESTED AREA FOR VEGETABLES 2021 - 2023 (B - SEASON)",
    size=11,
    pad=20,
    fontweight="bold",
    color="xkcd:dark blue",
)

axes[0].set_xticks(x)
axes[0].set_xticklabels(seasons_ordered)
axes[0].legend()

# ANALYSIS OF THE AREAS (Cultivated Crop area (Ha), Harvested Crop area (Ha)) FOR EACH SEASON - Fruits
# CULTIVATED AREA (Ha)
axes[1].bar(
    x - width / 2,
    veggies_and_fruits.iloc[1, 0:3],
    width,
    label="CULTIVATED AREA (Ha)",
    color="xkcd:light pink",
    linewidth=1,
    edgecolor="xkcd:dark blue",
)
axes[1].bar_label(
    axes[1].containers[0], label_type="center", color="xkcd:dark blue", fontsize=12
)

# HARVESTED AREA (Ha)
axes[1].bar(
    x + width / 2,
    veggies_and_fruits.iloc[1, 3:6],
    width,
    label="HARVESTED AREA (Ha)",
    color="xkcd:light yellow",
    linewidth=1,
    edgecolor="xkcd:dark blue",
)
axes[1].bar_label(
    axes[1].containers[1], label_type="center", color="xkcd:dark blue", fontsize=12
)

axes[1].set_ylabel(
    "Area (Ha)", size=12, labelpad=8, fontweight="bold", color="xkcd:dark blue"
)
axes[1].set_title(
    "CULTIVATED AREA VS HARVESTED AREA FOR FRUITS 2021 - 2023 (B - SEASON)",
    size=11,
    pad=20,
    fontweight="bold",
    color="xkcd:dark blue",
)

axes[1].set_xticks(x)
axes[1].set_xticklabels(seasons_ordered)
axes[1].legend()

plt.show()
```

```python
# ANALYSIS OF THE YIELD (Yield (MT/ha)) FOR EACH SEASON - Vegetables vs Fruits EACH SEASON
fig, axes = plt.subplots(
    1,
    3,
    figsize=(20, 5),
    sharey=True,
    dpi=100,
    facecolor="xkcd:light grey",
    edgecolor="xkcd:dark blue",
)

x = np.arange(2)
width = 0.4

for i in range(3):
    axes[i].pie(
        veggies_and_fruits.iloc[:, i + 9],
        labels=veggies_and_fruits.index,
        autopct="%1.1f%%",
        shadow=True,
        startangle=90,
        colors=["green", "orange"],
    )

    axes[i].set_title(
        seasons_ordered[i], size=13, fontweight="bold", color="xkcd:dark blue", pad=20
    )
    axes[i].axis("equal")

fig.suptitle(
    "VEGETABLES AND FRUITS YIELD 2021 - 2023 (B - SEASON)",
    size=15,
    fontweight="bold",
    color="xkcd:dark blue",
    y=1.1,
)
plt.show()
```
